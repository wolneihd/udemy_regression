```python
import pandas as pd
import numpy as np
import seaborn as sns

import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
diamonds = pd.read_csv('Course Materials/Data/Diamonds Prices2022.csv')
```

```python
diamonds.info()
```

```python
diamonds.corr(numeric_only=True)
```

```python
# feature
X = sm.add_constant(diamonds[['carat','x']])

# y target
y = diamonds['price']

model = sm.OLS(y, X).fit()

model.summary()
```

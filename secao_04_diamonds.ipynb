```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = pd.read_csv('Course Materials/Data/Diamonds Prices2022.csv')
```

```python
sns.histplot(diamonds['price'])
```

```python
sns.boxplot(x=diamonds['price'])
sns.despine()
```

```python
sns.histplot(diamonds['carat'])
plt.grid(ls=':')
```

```python
diamonds['cut'].value_counts().plot.barh()
```
